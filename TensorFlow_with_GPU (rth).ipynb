```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
```

```python
penguins = pd.read_csv('w.csv')
penguins.head()
```

```python
penguins = penguins.dropna()
penguins.head()
```

```python
import umap
```

```python
reducer = umap.UMAP()
```

```python
penguin_data0 = penguins[
    [
        "a1","a2","a3","a4",
        "a5","a6","a7","a8",
        "a9","a10",
        "a11","a12","a13","a14",
        "a15","a16","a17","a18",
        "a19","a20",
    ]
].values
scaled_penguin_data0 = StandardScaler().fit_transform(penguin_data0)
```

```python
embedding0 = reducer.fit_transform(scaled_penguin_data0)
embedding0.shape
```

```python
plt.scatter(
    embedding0[:, 0],
    embedding0[:, 1]
    # c=[sns.color_palette()[x] for x in penguins.species.map({"Adelie":0, "Chinstrap":1, "Gentoo":2})]
    )
c=[i for i in range(0, 24)]
for i, txt in enumerate(c):
    plt.annotate(txt, (embedding0[:, 0][i], embedding0[:, 1][i]))
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24);
```
