```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
```

```python
penguins = pd.read_csv('w.csv')
penguins.head()
penguins = penguins.dropna()
penguins.head()
```

```python
import umap
```

```python
reducer = umap.UMAP()
```

```python
penguin_data2 = penguins[
    [
        "a1","a2","a3","a4",
        "a5","a6","a7","a8",
        "a9","a10",
        "a11","a12","a13","a14",
        "a15","a16","a17","a18",
        "a19","a20",
        "a21","a22","a23","a24",
        "a25","a26","a27","a28",
        "a29","a30",
        "a31","a32","a33","a34",
        "a35","a36","a37","a38",
        "a39","a40",
    ]
].values
scaled_penguin_data2 = StandardScaler().fit_transform(penguin_data2)
```

```python
v1 = scaled_penguin_data2[:,:20]
v2 = scaled_penguin_data2[:,10:30]
v3 = scaled_penguin_data2[:,20:40]
vT = scaled_penguin_data2[:,:30]
```

```python
e1 = reducer.fit_transform(v1)
e2 = reducer.fit_transform(v2)
e3 = reducer.fit_transform(v3)
eT = reducer.fit_transform(vT)
```

```python
a = eT
plt.scatter(
    a[:, 0],
    a[:, 1]
    # c=[sns.color_palette()[x] for x in penguins.species.map({"Adelie":0, "Chinstrap":1, "Gentoo":2})]
    )
c=[i for i in range(0, 24)]
for i, txt in enumerate(c):
    plt.annotate(txt, (a[:, 0][i], a[:, 1][i]))
plt.gca().set_aspect('equal', 'datalim')
plt.title('Total', fontsize=24);
```
