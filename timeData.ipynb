```python
n_neighbors_x = 0
min_dist_x = 0
n_components_x = 2
metric_x = 0
premitive_data_size = 750
dimension_batch_size = 750
shard_batch_size = 0
cluster_size = 2
we = 1000


import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import umap
%matplotlib inline
df = pd.read_csv('fashion-mnist_test.csv')
print('The shape of testing dataset : ', df.shape)
df = df.loc[df['label'].isin([0,1])]
df['label']
df = df.dropna()
df.label.value_counts()
sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
vals = df.drop('label', axis=1).values
```

```python
v1 = vals[:50,:]
v2 = vals[25:75,:]
v3 = vals[50:100,:]
```

```python
reducer = umap.UMAP(n_components=n_components_x)
```

```python
emb1 = reducer.fit_transform(v1)
emb2 = reducer.fit_transform(v2)
emb3 = reducer.fit_transform(v3)
```

```python
plt.scatter(
    emb1[:, 0],
    emb1[:, 1],
    c=[sns.color_palette()[x] for x in df.label[:50].map({0:0, 1:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24)
plt.annotate("1", (emb1[:, 0][49], emb1[:, 1][49]))
plt.annotate("2", (emb1[:, 0][48], emb1[:, 1][48]))
plt.annotate("3", (emb1[:, 0][47], emb1[:, 1][47]))

plt.annotate("4", (emb1[:, 0][46], emb1[:, 1][46]))
```

```python
plt.scatter(
    emb2[:, 0],
    emb2[:, 1],
    c=[sns.color_palette()[x] for x in df.label[25:75].map({0:0, 1:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24);
plt.annotate("1", (emb2[:, 0][24], emb2[:, 1][24]))
plt.annotate("2", (emb2[:, 0][23], emb2[:, 1][23]))
plt.annotate("3", (emb2[:, 0][22], emb2[:, 1][22]))

plt.annotate("4", (emb2[:, 0][21], emb2[:, 1][21]))
```
